```python
import pandas as p
import numpy as np
```

```python
df = p.read_csv("Marks.csv")
print(df)
```

```python
df["total_marks"] = df.Subject1 +df.Subject2 +df.Subject3
```

```python
df
```

```python
avg = df.total_marks/3

df["avg"] =avg
df = df.astype({'avg':'int'})
def my_grade(df):
    if df.avg>=90:
        return "A" 
    elif(df.avg>=90 and df.avg<90):
        return "B"
    elif(df.avg>=70 and df.avg<80):
        return "C"
    elif(df.avg>=60 and df.avg<70):
        return " D"
    elif(df.avg>=50 and df.avg<60):
        return " E"
    else:
        return " F"

print()
df["Grade"]= df.apply(my_grade,axis=1)
```

```python
df
```

```python
df["Pass Or Fail"]="Pass"
df.loc[(df.Grade == 'F' ) | ( df.avg <= 50), 'Pass Or Fail'] = "Fail"
```

```python
df
```

```python
df.to_csv('new_marks.csv')
```
